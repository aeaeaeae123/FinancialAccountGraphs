```python
# plot_demo.py

import numpy as np
import matplotlib.pyplot as plt

def main():
    # Generate data
    x = np.linspace(0, 2 * np.pi, 200)
    y = np.sin(x)

    # Create the plot
    plt.figure(figsize=(6,4))
    plt.plot(x, y, linewidth=2)
    plt.title("Sine Wave")
    plt.xlabel("x")
    plt.ylabel("sin(x)")
    plt.grid(True)

    # Save to file so you can see it in the repo
    plt.savefig("sine_wave.png", dpi=150)
    plt.show()

if __name__ == "__main__":
    main()
```
