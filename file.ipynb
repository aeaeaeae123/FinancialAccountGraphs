## Distribution of Market Capitalization

```python
# %%
# 1) Declare and load the CSV file
import pandas as pd

# Update the path if your CSV lives elsewhere
csv_file = "./dataset/Finance_Accounting_Dataset.csv"
df = pd.read_csv(csv_file)

# Peek at the first few rows to confirm load
df.head()

# %%
# 2) Plot histogram of Market_Cap
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df["Market_Cap"], bins=50, edgecolor="black")
plt.title("Distribution of Market Capitalization")
plt.xlabel("Market Cap")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## 2. Box-plot of Profit by Industry

```python
# %%
# 1) Declare and load the CSV file (if not already loaded)
import pandas as pd

csv_file = "./dataset/Finance_Accounting_Dataset.csv"
df = pd.read_csv(csv_file)

# Peek at the data to confirm
df.head()

# %%
# 2) Box-plot of Profit by Industry
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x="Industry", y="Profit", data=df)
plt.title("Profit Distribution by Industry")
plt.xlabel("Industry")
plt.ylabel("Profit")
plt.xticks(rotation=45)   # Rotate industry labels for readability
plt.tight_layout()
plt.show()
```
